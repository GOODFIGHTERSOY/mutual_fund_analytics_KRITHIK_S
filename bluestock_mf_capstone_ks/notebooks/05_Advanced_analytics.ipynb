```python
import pandas as pd

# Load result files
var = pd.read_csv(r"C:\Users\KRITHIK\Downloads\bluestock_mf_capstone_ks\data\processed\var_cvar_report.csv")
cohort = pd.read_csv(r"C:\Users\KRITHIK\Downloads\bluestock_mf_capstone_ks\data\processed\cohort_analysis.csv")
sip = pd.read_csv(r"C:\Users\KRITHIK\Downloads\bluestock_mf_capstone_ks\data\processed\sip_continuity.csv")
hhi = pd.read_csv(r"C:\Users\KRITHIK\Downloads\bluestock_mf_capstone_ks\data\processed\sector_hhi.csv")

# Highest VaR
highest_var = var.loc[var['VaR_95'].idxmin()]

# Highest CVaR
highest_cvar = var.loc[var['CVaR_95'].idxmin()]

# Highest total investment cohort
highest_cohort = cohort.loc[cohort['Total_Investment'].idxmax()]

# Number of At-Risk investors
at_risk = sip[sip['Investor_Status'] == 'At-Risk']

# Highest HHI
highest_hhi = hhi.loc[hhi['HHI'].idxmax()]

print("\n----------- ADVANCED ANALYTICS SUMMARY -----------\n")

print(f"1. Highest VaR Fund")
print(f"   AMFI Code : {highest_var['amfi_code']}")
print(f"   VaR (95%) : {highest_var['VaR_95']:.4f}\n")

print(f"2. Highest CVaR Fund")
print(f"   AMFI Code : {highest_cvar['amfi_code']}")
print(f"   CVaR (95%): {highest_cvar['CVaR_95']:.4f}\n")

print(f"3. Highest Investing Cohort")
print(f"   Cohort : {highest_cohort['Cohort']}")
print(f"   Total Investment : {highest_cohort['Total_Investment']:.2f}\n")

print(f"4. At-Risk Investors")
print(f"   Total At-Risk Investors : {len(at_risk)}\n")

print(f"5. Highest Sector Concentration")
print(f"   AMFI Code : {highest_hhi['amfi_code']}")
print(f"   HHI : {highest_hhi['HHI']:.2f}")
```
